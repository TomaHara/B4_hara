```python
import pandas as pd

df = pd.read_csv("/Users/toma/Dropbox/all_sleep_data.csv")

print(df.columns)

df = df[df['type'] == 'long_sleep']

display(df[["time_in_bed", "type"]])

df = df[['user', 'average_breath', 'average_heart_rate', 'average_hrv', 'lowest_heart_rate',
       'time_in_bed','awake_time', 'deep_sleep_duration', 'light_sleep_duration', 'rem_sleep_duration', 'total_sleep_duration',
       'efficiency', 'latency', 
       'score']]

"""df = df[['user','awake_time', 
       'deep_sleep_duration', 'light_sleep_duration',
       'time_in_bed', 'total_sleep_duration']]"""

display(df)

#df = df.dropna()

display(df)

print(df.dtypes)
```

```python
from sklearn.preprocessing import StandardScaler

# ユーザーごとの平均を計算
grouped_df = df.groupby('user').mean().reset_index()

# 結果を表示
display(grouped_df)

print(grouped_df["user"].unique())

#　データフレームを標準化
col = ['average_breath', 'average_heart_rate', 'average_hrv', 'lowest_heart_rate',
       'time_in_bed','awake_time', 'deep_sleep_duration', 'light_sleep_duration', 'rem_sleep_duration', 'total_sleep_duration',
       'efficiency', 'latency', 
       'score']
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(grouped_df[col]), columns=col)
# 'user' 列を元のデータフレームからコピーし、最初の列に移動
scaled_df.insert(0, 'user', grouped_df['user'])

display(scaled_df)

display(scaled_df.describe())
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap.umap_ as umap
import matplotlib.pyplot as plt
import seaborn as sns


# ユーザー列を削除して次元削減に使用するデータを準備
features = scaled_df.drop('user', axis=1)

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(features)

# t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(features)

# UMAP
umap_reducer = umap.UMAP(n_components=2, random_state=42)
umap_result = umap_reducer.fit_transform(features)

# 可視化
fig, axes = plt.subplots(1, 3, figsize=(24, 8))

# PCAプロット
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=scaled_df['user'], ax=axes[0], palette='viridis')
axes[0].set_title('PCA')

# t-SNEプロット
sns.scatterplot(x=tsne_result[:, 0], y=tsne_result[:, 1], hue=scaled_df['user'], ax=axes[1], palette='viridis')
axes[1].set_title('t-SNE')

# UMAPプロット
sns.scatterplot(x=umap_result[:, 0], y=umap_result[:, 1], hue=scaled_df['user'], ax=axes[2], palette='viridis')
axes[2].set_title('UMAP')

plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap.umap_ as umap
import matplotlib.pyplot as plt
import seaborn as sns

# ユーザー列を削除して次元削減に使用するデータを準備
features = scaled_df.drop('user', axis=1)

# PCA
pca = PCA(n_components=3)  # 3次元に設定
pca_result = pca.fit_transform(features)

# t-SNE
tsne = TSNE(n_components=3, random_state=42)  # 3次元に設定
tsne_result = tsne.fit_transform(features)

# UMAP
umap_reducer = umap.UMAP(n_components=3, random_state=42)  # 3次元に設定
umap_result = umap_reducer.fit_transform(features)

# 可視化
fig = plt.figure(figsize=(24, 8))

# PCAプロット
ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(pca_result[:, 0], pca_result[:, 1], pca_result[:, 2], c=scaled_df['user'], cmap='viridis')
ax1.set_title('PCA')

# t-SNEプロット
ax2 = fig.add_subplot(132, projection='3d')
ax2.scatter(tsne_result[:, 0], tsne_result[:, 1], tsne_result[:, 2], c=scaled_df['user'], cmap='viridis')
ax2.set_title('t-SNE')

# UMAPプロット
ax3 = fig.add_subplot(133, projection='3d')
ax3.scatter(umap_result[:, 0], umap_result[:, 1], umap_result[:, 2], c=scaled_df['user'], cmap='viridis')
ax3.set_title('UMAP')

plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap.umap_ as umap
import matplotlib.pyplot as plt
import seaborn as sns

# ユーザー列を削除して次元削減に使用するデータを準備
features = grouped_df.drop('user', axis=1)

fig, axes = plt.subplots(3, 9, figsize=(27, 9))

# PCA with different random states
for i, ax in enumerate(axes[0]):
    random_state = i
    pca = PCA(n_components=2, random_state=random_state)
    pca_result = pca.fit_transform(features)
    
    sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=grouped_df['user'], ax=ax, palette='viridis')
    ax.set_title(f'PCA random_state={random_state}')

# t-SNE with different random states
for i, ax in enumerate(axes[1]):
    random_state = i
    tsne = TSNE(n_components=2, random_state=random_state)
    tsne_result = tsne.fit_transform(features)
    
    sns.scatterplot(x=tsne_result[:, 0], y=tsne_result[:, 1], hue=grouped_df['user'], ax=ax, palette='viridis')
    ax.set_title(f't-SNE random_state={random_state}')

# UMAP with different random states
for i, ax in enumerate(axes[2]):
    random_state = i
    umap_reducer = umap.UMAP(n_components=2, random_state=random_state)
    umap_result = umap_reducer.fit_transform(features)
    
    sns.scatterplot(x=umap_result[:, 0], y=umap_result[:, 1], hue=grouped_df['user'], ax=ax, palette='viridis')
    ax.set_title(f'UMAP random_state={random_state}')

plt.tight_layout()
plt.show()
```
